# Exercise 1


A matrix, B, is defined by the column vectors (8,0,0), (0,10,0), (0, 0, 4) \

1. What type of matrix would you call this?

2. Calculate the trace(B). \

3. Calculate det(B) \

4. Calculate the norm (L2) of B using both a python function and "manually" in python (but without using a built-in python function - how would you do that?) The results should match.

5. Given a vector, v, defined by the coordinates (2,3,4), calculate Bv.



```python
import numpy as np

B = np.array([[8, 0, 0],
              [0, 10, 0],
              [0, 0, 4]])
print ("1. What type of matrix would you call this?")
print ("Svar: Da 8, 10 og 4 sidder diagonelt og resten er 0, er det en diagonal matrix","\n")

print ("2. Calculate the trace(B)")
print (np.trace(B), "\n")

print ("3. Calculate det(B)")
print (np.linalg.det(B), "\n")

print ("4. Calculate the norm (L2) of B using both a python function and manually in python (but without using a built-in python function - how would you do that?) The             results should match.")
L2 = (8**2 + 10**2 + 4**2) ** 0.5
print ("L2 beregnet uden python funktioner: ",L2)
print ("Test med python funktion: ",np.linalg.norm(B), "\n")

print ("5. Given a vector, v, defined by the coordinates (2,3,4), Calculate Bv")

v = np.array([2,3,4])

Bv = B @ v

print (Bv, "\n")
```

# Exercise 2

Given a matrix, A, defined by the column vectors (1,3,5), (2,5,9), (2,3,6) and (-6,-9,-18).

1. Construct the matrix A. \
2. Determine the rank of A. \
3. Determine det(A).  \

4. Are all the column vectors linearly independent? \
5. Are all the row vectors linearly independent? \

6. Does the vector (1,2,3,5) belong to the nullspace of A? \
7. What is the dimension of the nullspace of A? \
8. Describe the form of all vectors in the nullspace of A. \



```python
import numpy as np
import scipy.linalg as la

A = np.array([[1,2,2,-6],
              [3,5,3,-9],
              [5,9,6,-18]])
print ("1. Construct the matrix A.")
print ("Matrix A:", "\n",A, "\n")

print ("2. Detemine the rank of A")
rank = np.linalg.matrix_rank(A)
print ("Rank af Matrix A: ",rank, "\n")

print ("3. Determine det(A)")
print("Eftersom matrixen ikke er kvadrat kan vi ikke finde det(A)", "\n")

print ("4. Are alle the collumn vectors linearly independent?")
print ("Eftersom vores rank er 3 og der er 4 kolonner, er den ikke lineært uafhængig","\n")

print ("5. Are all the row vectors linearly independent?")
print ("Ja det er de, eftersom der er 3 række vectorer, og vores rank er 3, er de lineært uafhængige","\n")

print ("6. Does the vector (1,2,3,5) belong to the nullspace of A?")
vector = A @ np.array([1,2,3,5])
print (vector)
if np.all(vector == 0):
    print("Ja, vektoren tilhører nullspacet\n")
else:
    print("Nej, vektoren tilhører ikke nullspacet\n")

print ("7. What is the dimension of the nullspace of A")
nullspace = la.null_space(A)
print (nullspace.shape, "\n")

print ("8. Describe the form of all vectors in the nullspace of A")
print("Alle vektorer i nullspacet er på formen x = t * n, hvor n er:")
print (np.round(nullspace, 2))
print ("og t er et vilkårligt reelt tal")


```

# Exercise 4

Just as there is a difference between rotation around the center of a figure and rotation around (0,0), the same applies to scaling.

\
Per default points are scaled "relative to" (0,0) and not the center of the figure.

But we can apply the same technique as we did with rotation around the center of the figure in the lectures, to apply scaling around the center of the figure.

Given a square defined by the points (2,2), (2,4), (4,4), (4,2)

1. Put the points in a matrix and visualize the points.

3. Then scale the figure by 2 around it's center. \

4. Visualize the new transformed figure.

```python
import numpy as np
import matplotlib.pyplot as plt

print ("1. Put the points in a matrix and visualize the points")
square = np.array ([[2,2],
                    [2,4],
                    [4,4],
                    [4,2]])
print (square)

plt.grid(True)
plt.axis('equal')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.spines['left'].set_position('zero')  # y-axis through x=0
ax.spines['bottom'].set_position('zero')  # x-axis through y=0
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.scatter(square [:,0], square[:,1],color='red')
plt.show()

print ("2. Then scale the figure by 2 around it's center")
scale = 2
center = np.mean(square, axis = 0)
square_scaled = (square - center) * scale + center
print (square_scaled)

print ("3. Visualize the new transformed figure")
plt.grid(True)
plt.axis('equal')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.spines['left'].set_position('zero')  # y-axis through x=0
ax.spines['bottom'].set_position('zero')  # x-axis through y=0
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.scatter(square_scaled [:,0], square_scaled[:,1],color='red')
plt.show()
```

Tetris is an old computer game where you have to rotate and translate figures. Take a look at the following figure. \


computer graphics - tetris example.png



1. Put the coordinates of the points of F1 into a matrix.

2. Visualize the F1 points (you do not need to draw the lines between points or the labels)

3. Figure out a combination of transformations (rotation, translation, scaling etc) that will transform F1 into F2.
(Note: Several different solutions can yield the same result)

4. Apply these transformations to the F1 matrix.

5. Visualize the resulting F2 figure and check it corresponds to the image.

```python
import numpy as np
import matplotlib.pyplot as plt

print ("1. Put the coordinates of the points of F1 into a matrix")
F1 = np.array ([[2,6],
                [2,4],
                [3,4],
                [3,5],
                [5,5],
                [5,6]])

print ("2. Visualize the F1 points (you do not need to draw the lines between points or the labels)")
origin=np.array([[0,0],[0,0]])
plt.grid(True)
plt.axis('equal')
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.spines['left'].set_position('zero')  # y-axis through x=0
ax.spines['bottom'].set_position('zero')  # x-axis through y=0
plt.xlim(-7, 7)
plt.ylim(-1, 7)
plt.scatter(F1 [:,0], F1[:,1],color='red')
plt.show()


print ("3. Figure out a combination of transformations (rotation, translation, scaling etc) that will transform F1 into F2")

print ("Vi roterer F1 90 grader med uret")
theta = -np.pi/2
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

print("4. Apply transformations to F1")
F1_rotated = np.round((R @ F1.T).T).astype(int)

print ("F2 skal ligge ved x=-5 til -3, y=1 til 4")
print ("Vi finder translation ud fra hvor figuren skal starte")
tx = -5 - F1_rotated[:, 0].min()
ty =  1 - F1_rotated[:, 1].min()
translation = np.array([tx, ty])
print("Translation:", translation)

F2 = F1_rotated + translation
print("F2 koordinater: \n", F2)

print("5. Visualize F2 and check it corresponds to the image")
plt.grid(True)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
plt.xlim(-7, 7)
plt.ylim(-1, 7)
plt.scatter(F1[:, 0], F1[:, 1], color='blue',  label='F1 (original)')
plt.scatter(F2[:, 0], F2[:, 1], color='red',   label='F2 (transformeret)')
plt.legend()
plt.show()

```
