```python
import numpy as np
import matplotlib.pyplot as plt

# used to create non-regular subplots
import matplotlib.gridspec as gridspec

Q,R = np.linalg.qr(np.random.rand(5,5))

print (np.round (Q))
print (np.round(Q.T @ Q)) # should be identity
print (np.round(Q @ Q.T))
print (np.round(np.linalg.inv (Q) @ Q))
print (np.round(Q @ (np.linalg.inv(Q))))

plt.imshow(Q)
plt.title("Q from QR decomposition")
plt.show()

plt.imshow(np.linalg.inv(Q))
plt.title("Q from QR decomposition")
plt.show()
```
