# Problem 3

Now, as we have seen, vectors from a given vector space can be represented by their coordinates with respect to a specific basis.

NumPy operates on these coordinates. It does not implement the abstract concepts of vector spaces, bases, or coordinate systems. Instead, it provides numerical operations on arrays of numbers. Any knowledge about the underlying vector space structure or the choice of basis in which vectors are represented is not encoded in the software, but must be supplied by the user.

So let us say we have two vectors in $\mathbb{R}^2$


\begin{array}{l}
\mathbf{v}_1 = (-1, 9) \\
\mathbf{v}_2 = (13, 5)
\end{array}
These vectors can be represented with coordinates
\begin{align*}
v_1&=(4,5)\\
v_2&=(9,-4)
\end{align*}
In the basis
\begin{align*}
\textbf{b}_1&=(1,1)\\
\textbf{b}_2&=(-1,1)
\end{align*}


Then we can represent the two vectors as numpy arrays in **that** basis by

    v_1=np.array([4,5])

    v_2=np.array([9,-4])

# Plotting Arrows
In the next exercise we are going to take a look at visualization

To use visualization tools you should import

```python

import matplotlib.pyplot as plt

```
This will give access  to plot facilities

We can, if we want to, represent vectors as arrows from the origin

For this purpose we can use the matplotlib.pyplot.quiver function

````python
import numpy as np
import matplotlib.pyplot as plt

v=np.array([[1,2],[2,3]])
origin=np.array([[0,0],[0,0]])


plt.grid(True)
plt.axis('equal')
ax = plt.gca() ## get current axis(gca)
ax.set_aspect('equal', adjustable='box')
ax.spines['left'].set_position('zero')  # y-axis through x=0
ax.spines['bottom'].set_position('zero')  # x-axis through y=0
plt.xlim(-1, 4) # x-limits
plt.ylim(-1, 4) #y-limits
plt.quiver(origin[:,0], origin[:,1], v[:,0], v[:,1], angles='xy', scale_units='xy', scale=1,
               color='red')

````

In $\mathbb{R}^4$ consider the two different bases
\begin{align*}
\mathbf{b}_1&=(1,1,1,1)\\
\mathbf{b}_2&=(1,1,-1,1)\\
\mathbf{b}_3&=(1,-1,1,1)\\
\mathbf{b}_4&=(1,1,1,-1)
\end{align*}

\begin{align*}
\mathbf{f}_1&=(1,-14,23,1)\\
\mathbf{f}_2&=(1,0,-1,1)\\
\mathbf{f}_3&=(1,-1,1,0)\\
\mathbf{f}_4&=(1,1,0,-1)
\end{align*}


First we have to decide whether these bases suggestions are actual bases that is, they are linear independent sets





````python
import numpy as np
#define all b basis vectors as 1-dim arrays
b_1 = np.array ([1,1,1,1])
b_2 = np.array ([1,1,-1,1])
b_3 = np.array ([1,-1,1,1])
b_4 = np.array ([1,1,1,-1])

#end

#Stack to matrix and assert det non-zero
b = np.column_stack((b_1,b_2,b_3,b_4))
print("b = ","\n",b)

detb = np.linalg.det(b)
print("Determinanten for b = ",detb)

assert detb != 0, "Determinanten er 0 og er ikke lineært uafhængige"
print ("Vektorerne udgør en gyldig basis, da de er non-zero")

#end

#define all f basis vectors as 1-dim arrays
f_1 = np.array ([1,-14,23,1])
f_2 = np.array ([1,0,-1,1])
f_3 = np.array ([1,-1,1,0])
f_4 = np.array ([1,1,0,-1])
#end

#Stack to matrix and assert det non-zero
f = np.column_stack((f_1,f_2,f_3,f_4))
print("f = ","\n",f)

detf = np.linalg.det(f)
print("Determinanten for f = ",detf)

assert detb != 0, "Determinaten er 0 og er ikke lineært uafhængige"
print ("Vektorerne udgør en gyldig basis, da de er non-zero")
#end
````

# Problem 7
Now consider vector  $v\in \mathbb{R}^4$ with coordinates $(3,5,8,-1)$ in basis $\mathbf{b}_i$. What are the coordinates in basis $\mathbf{f}_i$

````python
import numpy as np
#define the coordinates as 1-dim array v_b
v_b = np.array ([3,5,8,-1])
# end

#Use the definitions of the transformation matrices from above to calculate the coordinates v_f in the f basis and print
b_1 = np.array([1,1,1,1])
b_2 = np.array([1,1,-1,1])
b_3 = np.array([1,-1,1,1])
b_4 = np.array([1,1,1,-1])
b = np.column_stack([b_1, b_2, b_3, b_4])


f_1 = np.array ([1,-14,23,1])
f_2 = np.array ([1,0,-1,1])
f_3 = np.array ([1,-1,1,0])
f_4 = np.array ([1,1,0,-1])
f = np.column_stack((f_1,f_2,f_3,f_4))


v_f = np.linalg.inv(f) @ b @ v_b
print("v_f = ",v_f)

#end
# Verify the result. This can be done by using np.allclose(a,b) which compares all data in the two arrays
# And then comparing B@v_b and F@v_f. why?
print("Verificeret = ",np.allclose(b @ v_b, f @ v_f))


````
